# Lightweight networks and MobileNet

We have seen that complex networks require significant computational resources for training and for fast inference. It turns out that a model with a significantly smaller number of parameters in most cases can still be trained to perform reasonably well. In the beginning of the module when training MNIST digit classification to illustarted this. The accuracy of a simple dense model was not significantly worse than that of a powerful CNN. Increasing the number of CNN layers and/or number of neurons in the classifier allowed us to gain a few percents of accuracy at most.

This leads us to the idea that we can experiment with a lightweight network architectures in order to train faster models. This is especially important if we want to be able to execute our models on mobile devices.

## MobileNet

In the previous unit, we have seen the **ResNet** architecture for image classification. A more lightweight analog of ResNet is **MobileNet**, which uses the *Inverted Residual Blocks*. Let's load the pre-trained mobilenet and see how it works.

> As before, we need to pre-load model weights in the sandbox environment. You can ignore the next cell if you are running the notebook locally.

```python
# Download model weights in the sandbox environment
!mkdir -p ~/.keras/models
!wget -P ~/.keras/models -q https://mslearntensorflowlp.blob.core.windows.net/models/mobilenet_v2_weights_tf_dim_ordering_tf_kernels_1.0_224.h5
```

```python
model = keras.applications.MobileNetV2()
model.summary()
```

Let's apply the model to our dataset and make sure that it works.

```python
x_sample = next(ds_train.as_numpy_iterator()) # get first batch of training data
image = x_sample[0][:1].copy() # get a copy of first image

inp = keras.applications.mobilenet.preprocess_input(image)

res = model(inp)
print(f"Most probable class = {tf.argmax(res,1)}")

keras.applications.mobilenet.decode_predictions(res.numpy())
```

## Using MobileNet for transfer learning

Now let's perform the same transfer learning process as in previous unit but using MobileNet. When using pre-trained networks it is essential to apply the pre-processing step to all input images. In our previous unit we did not do that but the model still trained fine. This was because VGG-16 is not very sensitive to pre-processing. If you tried to adjust the code from previous lesson to use MobileNet, you would not be able to get the model to train.

One of the ways to add pre-processing is to apply pre-processing steps to the original dataset like this:

```python
ds_train_pre = ds_train.map(lambda x,y : (keras.applications.mobilenet.preprocess_input(x),y))
ds_test_pre = ds_test.map(lambda x,y : (keras.applications.mobilenet.preprocess_input(x),y))
```

We can also do something even more clever by adding a preprocessing layer directly to our neural network!

## Functional Network Definition

To define a neural network that includes some complex computations it is convenient to use another syntax for defining a neural network computational graph called the **Functional definition**. It starts by definining a neural network input variable indicating its shape:

```python
inp = keras.Input(shape=(224,224,3))
```

We will also define a mobilenet network and make it non-tranable:

```python
mob = keras.applications.MobileNetV2(include_top=False,weights='imagenet',input_shape=(224,224,3))
mob.trainable = False
```

Then, we can perform a number of operations on this input, in order to produce some output, i.e. provide a Python code to compute the output using all available operations:

```python
x = keras.applications.mobilenet.preprocess_input(inp)
x = mob(x)
x = keras.layers.GlobalAveragePooling2D()(x)
out = keras.layers.Dense(1,activation='sigmoid')(x)
```

Finally, we define a model just by giving it input and output variables:

```python
model = keras.models.Model(inp,out)
model.summary()
```

A functional way of defining a model turns out to be very useful when we have complex models which cannot be represented just by a composition of layers.

It is interesting to observe the structure of our model. We can see that `Input` is represented by a separate layer. Also, `preprocess_input` function is represented by two layers, division and subtraction (because that is how normalization is done). 

To minimize the number of parameters in the dense layer, we will have also applied **Gloval Average Pooling** layer after the convolutional base. It is a trick that is quite often used in transfer learning.

Let's now see how the model trains:

```python
model.compile(loss='binary_crossentropy',metrics=['acc'],optimizer='adam')
hist = model.fit(ds_train,validation_data=ds_test)
```

## Takeaway

By carefully architecting the MobileNet-based model we were able to achieve the best results so far in the cats vs. dogs problem. While training a model based on ResNet would have probably given us even better accuracy the difference is not that big. The MobileNet model is significanlty smaller and faster during inference, and can even be used on CPU devices.

One of the important advantages of small models, such as MobileNet or ResNet-18, is that they can be used on mobile devices. Specifically for mobile devices, we can also use [Tensorflow Lite](https://www.tensorflow.org/lite). The technique is often used is to **convert** the trained model into a mobile version, which can then be executed on iOS/Android mobile device, or on embedded a microcontroller.
